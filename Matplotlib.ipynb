## Plots

### Line graph

```python
# set figsize
plt.rcParams['figure.figsize'] = (8,4)

# x and y list axis
x = [1, 2, 6, 8, 12, 18]
x2 = np.arange(0,12,0.5)
y = [5, 20, 10, 75, 55, 100]

# Different types of graphs + work with [color][marker][line] format
plt.plot(x,y, 'go:', label = 'A simple line', linewidth = 2, markersize=5, markeredgecolor='white') # dotted plot
plt.plot(x2[:10], x2[:10]**2, 'r', label='x^2.1')    # plot with line
plt.plot(x2[9:], x2[9:]**2, 'b--', label='x^2.2')      # continiue plot dashed line

# adding titles to graph and axis
plt.title('Line Graph', loc = 'center', fontdict = {'fontname': 'Times New Roman', 'color': 'white', 'fontsize' : '24'})
plt.xlabel('X-axis', fontdict = {'color': 'white', 'fontsize' : '12'})
plt.ylabel('Y-axis', fontdict = {'color': 'white', 'fontsize' : '12'})
     
# adding legend
plt.legend(loc = 0)

# save and show graph
plt.savefig('grapg1.png', dpi = 250)
plt.show()
```

## Three Plots Line-Graph Screen

```python
x = np.linspace(0, 20, 20)
y = np.random.randint(1, 100, 20)

# Set figure and axes
fig, ax = plt.subplots(1,3, figsize=(14, 5))
ax[0].plot(x,y, 'gp--', linewidth = 2, markersize=5, markeredgecolor='yellow')
ax[1].plot(x, y*x, 'c8:', linewidth = 2, markersize=5, markeredgecolor='white')
ax[2].plot(x, y*y, 'r*:', linewidth = 2, markersize=5, markeredgecolor='cyan')

# Set titels
fig.suptitle("Three Plots Screen",  fontdict = {'fontname': 'Calibri', 'color': 'white'})
ax[0].set_title("plot 1 (x,y)", fontdict = {'fontname': 'Calibri', 'color': 'white'})
ax[1].set_title("plot 2 (x, x*y)", fontdict = {'fontname': 'Calibri', 'color': 'white'})
ax[2].set_title("plot 3 (x, y*y)", fontdict = {'fontname': 'Calibri', 'color': 'white'})

plt.savefig('ThreePlots.png')
plt.show()
```

## Multiple 6 Plots in the same Canvas : Line-Graph

## Bar chart

```python
plt.rcParams['figure.figsize'] = (10,4)

x = [2, 4, 6, 8, 10]
y = [9, 6, 4, 7, 3]

# Set a new bar graph
bars = plt.bar(x, y, align = 'edge', edgecolor = 'w')

# Set colors to bars
bars[0].set_facecolor('y')
bars[1].set_facecolor('b')
bars[2].set_facecolor('c')
bars[3].set_facecolor('r')
bars[4].set_facecolor('g')

# Set hatches to bars
bars[0].set_hatch('/')
bars[1].set_hatch('-')
bars[2].set_hatch('o')
bars[3].set_hatch('+')
bars[4].set_hatch('*')

# Set titels
plt.title('Bar Graph', loc = 'left', fontdict = {'fontname': 'Times New Roman', 'color': 'white', 'fontsize': '30'})
plt.xlabel('X-axis', fontdict = {'color': 'white','fontsize' : '14'})
plt.ylabel('Y-axis', fontdict = {'color': 'white','fontsize' : '14'})

# rotate the y-axis
plt.yticks(rotation=30)

plt.savefig('Barchart1.png')
plt.show()
```

## 3-Bars Chart

### This chart comparing salarys between 3 types of engineers:
* Indastrial & Management Engineer
* Software Engineer
* Electrical Engineer

```python
plt.rcParams['figure.figsize'] = (10,6)

# loop for the last 15 years
years = [] 
l = 2008 
for i in range(15):
    l += 1
    years.append(l)

# Random function for 3 list - salarys of engineers    
indastrialSalary = np.random.randint(8000, 13000, 15)
softwareSalary = np.random.randint(15000, 25000, 15)
electricalSalary = np.random.randint(13000, 19000, 15)

# Convert list from string to int
x = np.arange(len(years))
plt.xticks(ticks=x, labels=years)

# Set a 3-bars in one graph 
plt.bar(x+0.3, indastrialSalary, color='r', label='Indastrial Salary', width=0.3)
plt.bar(x-0.3, softwareSalary, color='b', label='Software Salary', width=0.3)
plt.bar(x, electricalSalary, color='y', label='Electrical Salary', width=0.3)

# Set titels
plt.title('Three-Bar Graph', loc = 'center', fontdict = {'fontname': 'Times New Roman', 'color': 'white', 'fontsize': '30'})
plt.xlabel('X-axis', fontdict = {'color': 'white','fontsize' : '15'})
plt.ylabel('Y-axis', fontdict = {'color': 'white','fontsize' : '15'})

# Set legend
plt.legend(loc = 1, fontsize="10")

plt.savefig('Barchart3.png')
plt.show()
```

## Pie chart

```python
plt.rcParams['figure.figsize'] = (10,8)

# Adding parameters
labels = ['A', 'B', 'C', 'D', 'E']
parts = [1, 2, 7, 3, 5]
explode = [0, 0.1, 0, 0.2, 0.1]
colors = ['#FFACAC', '#FF7474', '#FF4B4B', '#FFC2C2','#D80000' ]

# Adding the pie chart
plt.pie(parts, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%', wedgeprops={'edgecolor': 'w', 'linewidth': 2})

# Legend + title
plt.title('Pie Chart', loc = 'center', fontdict = {'fontname': 'Times New Roman', 'color': 'white', 'fontsize': '30'})
plt.legend(labels)

plt.savefig('Piechart1.png')
plt.show()
```

## Stackplot Chart

#### On this page we will see the growth of the population in the world over the years in relation to continents

```python
plt.rcParams['figure.figsize'] = (10,4)

# Adding parameters
year = [1950, 1960, 1970, 1980, 1990, 2000, 2010, 2018]

africa = [219, 275, 354, 465, 620, 801, 1031, 1262]
america = [327, 412, 508, 605, 715, 825, 932, 1006]
asia = [1273, 1574, 2008, 2510, 3090, 3602, 4054, 4448]
europe = [207, 240, 264, 283, 298, 291, 280, 281]

# Adding stackplot
plt.stackplot(year, africa, america, asia, europe)

# Legend + title
plt.title('Population', loc = 'center', fontdict = {'fontname': 'Times New Roman', 'color': 'white', 'fontsize': '30'})
plt.legend(['Africa', 'America', 'Asia', 'Europe'], loc='upper left')

plt.savefig('Stckplot1.png')
plt.show()
```

##  Multiple 3 Plots in the same Canvas : Histograma

#### In this graph we will drawn a hundred numbers from 1 to 100, and check:

* Chart 1: How many numbers fell for each number between 1-9?
* Chart 2: How many numbers fell between three central ranges?
* Chart 3: How many numbers fell within the ranges that i set in freestyle?

```python
# how much numbers between 1-9 randomaly drawn from 100 numbers ?

# nums randomaly function
numbers = np.random.randint(1,9,100)

# Set figure and axes
fig, ax = plt.subplots(1,3, figsize=(14, 5))
ax[0].hist(numbers, bins=10, edgecolor = 'w', facecolor='#C28F1E') # between 1-9 
ax[1].hist(numbers, bins=3, edgecolor = 'w') # between 3-range of bins groups
ax[2].hist(numbers, bins=[0, 2, 5, 7, 9], edgecolor = 'w', facecolor='#C063F5') # between 4-range bins groups (my definition)

# set mean line to the center chart
mean = sum(numbers) / len(numbers)
ax[1].axvline(mean, color='red', label='Mean')

# Set titels
fig.suptitle("Three Histograma Screen",  fontdict = {'fontname': 'Times New Roman', 'color': 'white',})
ax[0].set_title("between 1-9", fontdict = {'fontname': 'Calibri', 'color': 'white', 'fontsize': '20'})
ax[1].set_title("3-range of bins", fontdict = {'fontname': 'Calibri', 'color': 'white', 'fontsize': '20'})
ax[2].set_title("4-range bins (my definition)", fontdict = {'fontname': 'Calibri', 'color': 'white', 'fontsize': '20'})

plt.savefig('Histograma1.png')
plt.show()
```

##  Multiple 3 Plots in the same Canvas : Scatter chart

* Chart 1: Parabula (x, x^2)
* Chart 2: Random numbers
* Chart 3: Polinom (x, x^3)

```python
plt.rcParams['figure.figsize'] = (10,4)

x = np.linspace(-10, 10, 20) 
y = np.random.randint(-30, 100, 20)

# Set figure and axes
fig, ax = plt.subplots(1,3, figsize=(14, 5))

# Adding new scatter charts
ax[0].scatter(x, x**2, s=100, c='r', marker='X') # parabula
ax[1].scatter(x, y, s=100, c='c', marker='3') # Random numbers
ax[2].scatter(x, x**3, s=100, c='g', marker='+') # x^3

# Set titels
fig.suptitle("Three Scatter Screen",  fontdict = {'fontname': 'Times New Roman', 'color': 'white',})
ax[0].set_title("Scatter 1 (x, x^2) ", fontdict = {'fontname': 'Calibri', 'color': 'white', 'fontsize': '20'})
ax[1].set_title("Scatter 2 (x, z)", fontdict = {'fontname': 'Calibri', 'color': 'white', 'fontsize': '20'})
ax[2].set_title("Scatter 3 (x, x^3)", fontdict = {'fontname': 'Calibri', 'color': 'white', 'fontsize': '20'})

# Set red line when y=0
ax[1].axhline(y=0, c='r', linestyle='--')
ax[2].axhline(y=0, c='r', linestyle='--')

plt.savefig('Scatter1.png')
plt.show()
```

* Strength of the color: will be the heat of the person's body

```python
# Loading a data that containing time and temperature of solar radiation
with open('small.txt', 'r') as f: data = np.genfromtxt(f, dtype='datetime64[s],f,f,f', names=['date', 'doy', 'temp', 'solar'])
datetime = data['date']
dayofyear = data['doy']
temperature = data['temp']
solar = data['solar']
temp_low = np.convolve(temperature, np.ones(48)/48, mode = 'same')
solar_low = np.convolve(solar, np.ones(48)/48, mode = 'same')
```

```python
fig, [ax0, ax1] = plt.subplots(1, 2, constrained_layout = True, figsize=(10, 6))
# background
ax0.set_facecolor('0.2')
ax1.set_facecolor('0.2')
# scatter
sc = ax0.scatter(solar, temperature, c = dayofyear, s = 2, cmap = 'autumn')
sc = ax1.scatter(solar_low[::24], temp_low[::24], c = dayofyear[::24], s = 9, cmap='autumn')
#labels
ax0.set_xlabel('Solar Radiation $[W\,m^{-2}]$')
ax1.set_xlabel('Solar Radiation $[W\,m^{-2}]$')
ax0.set_ylabel('Temperature $[^oC]$')
ax1.set_ylabel('Temperature $[^oC]$')
ax0.set_title('Hourly', loc = 'center')
ax1.set_title('Smoothed', loc = 'center')
fig.colorbar(sc, ax = [ax0, ax1], shrink = 0.6, label = 'day of year') # colorbar
```
